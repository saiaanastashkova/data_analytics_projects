## Приоретизация гипотез

Применим фреймворки ICE и RICE для приоритизации гипотез. Посмотрим изменится ли приоритизация гипотез при применении RICE вместо ICE.

### Фреймворк ICE

```python
hypothesis['ICE'] = round(((hypothesis['impact']*hypothesis['confidence'])/hypothesis['efforts']), 2)
print(hypothesis[['hypothesis','ICE']].sort_values(by='ICE', ascending=False))
```

При применении фреймворка ICE самыми приоритетными выдвигаются гипотезы:

1 место - Гипотеза № 9:
- Запустить акцию, дающую скидку на товар в день рождения

2 место - Гипотеза №1:
- Добавить два новых канала привлечения трафика, что позволит привлекать на 30% больше пользователей

3 место - Гипотеза №8:
- Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок

### Фреймворк RICE

```python
hypothesis['RICE'] = round(((hypothesis['reach']*hypothesis['impact']*hypothesis['confidence'])/hypothesis['efforts']), 2)
print(hypothesis[['hypothesis','RICE']].sort_values(by='RICE', ascending=False))
```

Картина поменялась при применении фреймворка RICE:

1 место - Гипотеза №8:
- Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок

2 место - Гипотеза №3:
- Добавить блоки рекомендаций товаров на сайт интернет магазина, чтобы повысить конверсию и средний чек заказа

3 место - Гипотеза №1:
- Добавить два новых канала привлечения трафика, что позволит привлекать на 30% больше пользователей

## A/B тестирование

Представим, что мы **выбрали для тестирования гипотезу №8 "Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок"**, провели A/B-тест и получили результаты, которые описаны в таблицах orders и visitors. 

Сначала проверим **не попадают ли какие-то пользователи в обе группы**. При наличии таких дубликатов посчитаем их долю от общего числа. Если их больше %, то будет целесообразно такие дубликаты удалить, чтобы они не искажали результаты теста.

```python
duplicated_users = orders.groupby('visitor_id').agg({'group':'nunique'}).reset_index()
duplicated_users = duplicated_users[duplicated_users['group']==2]
print(duplicated_users['visitor_id'].count())
print(duplicated_users['visitor_id'].count()/orders['visitor_id'].nunique())
```

Количество пользователей, попавших в обе группы, составляет 58 чел. или 5,6% от общего числа уникальных пользователей. Эта доля является допустимой, поэтому можем их оставить. 

Сгруппируем таблицу orders по дате и группе A/B-теста и посчитаем количество уникальных заказов, покупателей и суммарную выручку.

```python
orders_agg= orders.groupby(
    ['date','group']).agg({'transaction_id':'nunique', 'visitor_id':'nunique', 'revenue':'sum'}).reset_index()
orders_agg.head(10)
```

Добавим по группам кумулятивные данные о заказах, покупателях и сохраним в столбцы transactions_cum, visitors_cum, revenue_cum.

```python
orders_agg[['transactions_cum', 'customers_cum', 'revenue_cum']] = pd.concat(
    [orders_agg[orders_agg['group'] == 'A'][['transaction_id', 'visitor_id', 'revenue']].cumsum(), \
     orders_agg[orders_agg['group'] == 'B'][['transaction_id', 'visitor_id', 'revenue']].cumsum()])
orders_agg.head(10)
```

Те же манипуляции проведем с таблицей о посетителях visitors.

```python
visitors_agg = visitors.groupby(['date','group']).agg({'visitors':'sum'}).reset_index()
visitors_agg['visitors_cum'] = pd.concat(
    [visitors_agg[visitors_agg['group'] == 'A']['visitors'].cumsum(), \
     visitors_agg[visitors_agg['group'] == 'B']['visitors'].cumsum()])
visitors_agg.head(10)
```

Объединим обе таблицы в одну по столбцу с датой.

```python
data = orders_agg.merge(visitors_agg, left_on=['date', 'group'], right_on=['date', 'group'])
data.rename(columns = {'transaction_id':'transactions_daily', 'visitor_id':'customers_daily'}, inplace = True)
data.head(10)
```

Структура полученной таблицы по столбцам:

- transactions_daily — количество заказов в выбранную дату;
- customers_daily - количество покупателей в выбранную дату;
- revenue — суммарная выручка в выбранную дату;
- transactions_cum — суммарное число заказов до выбранной даты включительно;
- customers_cum - суммарное число покупателей до выбранной даты включительно;
- revenue_cum — суммарная выручка до выбранной даты включительно;
- visitors — количество пользователей в выбранную дату;
- visitors_cum  — количество пользователей до выбранной даты включительно.

### Выручка по группам 

Построим график кумулятивной выручки по группам. 

```python
# Таблица с кумулятивным значением количества заказов и выручки по дням в группах А и B
data_a = data[data['group']=='A'][['date','revenue_cum', 'transactions_cum']]
data_b = data[data['group']=='B'][['date','revenue_cum', 'transactions_cum']]

# Формируем графики выручки группы А и B
plt.figure(figsize=(10, 5))
plt.grid(True)
plt.title('График кумулятивной выручки по дням и группам A/B-тестирования')

plt.plot(data_a['date'], data_a['revenue_cum'], label='A')
plt.plot(data_b['date'], data_b['revenue_cum'], label='B')

ax = plt.gca()
ax.grid(True)
ax.set_xlabel('Дни заказов')
ax.set_ylabel('Выручка заказов')
plt.legend()
```

### Статистическая значимость различий в среднем количестве заказов на посетителя по «сырым» данным

Подготовим сводную таблицу для анализа.

```python
#Формируем таблицу с количеством посетителей группы A в каждый день и с кумулятивным количеством 
visitors_a_daily = visitors[visitors['group']=='A'][['date', 'visitors']]
visitors_a_daily.columns = ['date', 'visitors_per_date_a']

visitors_a_cum = visitors_agg[visitors_agg['group'] == 'A'][['date', 'visitors_cum']]
visitors_a_cum.columns = ['date', 'visitors_a_cum']
```

```python
#Формируем таблицу с количеством посетителей группы B в каждый день и с кумулятивным количеством 
visitors_b_daily = visitors[visitors['group']=='B'][['date', 'visitors']]
visitors_b_daily.columns = ['date', 'visitors_per_date_b']

visitors_b_cum = visitors_agg[visitors_agg['group'] == 'B'][['date', 'visitors_cum']]
visitors_b_cum.columns = ['date', 'visitors_b_cum']
```

```python
#Формируем таблицу с количеством заказов группы A в каждый день и с кумулятивным количеством 
orders_a_daily = orders[orders['group']=='A'][['date', 'transaction_id', 'visitor_id', 'revenue']]\
    .groupby('date', as_index=False)\
    .agg({'transaction_id' : pd.Series.nunique, 'revenue' : 'sum'})
orders_a_daily.columns = ['date', 'orders_per_date_a', 'revenue_per_date_a']

orders_a_cum = orders_agg[orders_agg['group'] == 'A'][['date', 'transactions_cum', 'revenue_cum']]
orders_a_cum.columns = ['date', 'orders_cum_a', 'revenue_cum_a']
```

```python
#Формируем таблицу с количеством заказов группы B в каждый день и с кумулятивным количеством 
orders_b_daily = orders[orders['group']=='B'][['date', 'transaction_id', 'visitor_id', 'revenue']]\
    .groupby('date', as_index=False)\
    .agg({'transaction_id' : pd.Series.nunique, 'revenue' : 'sum'})
orders_b_daily.columns = ['date', 'orders_per_date_b', 'revenue_per_date_b']

orders_b_cum = orders_agg[orders_agg['group'] == 'B'][['date', 'transactions_cum', 'revenue_cum']]
orders_b_cum.columns = ['date', 'orders_cum_b', 'revenue_cum_b']
```

```python
#Соединяем ранее сделанные таблицы в один датасет
data_total = orders_a_daily.merge(orders_b_daily, left_on='date', right_on='date', how='left')\
    .merge(orders_a_cum, left_on='date', right_on='date', how='left')\
    .merge(orders_b_cum, left_on='date', right_on='date', how='left')\
    .merge(visitors_a_daily, left_on='date', right_on='date', how='left')\
    .merge(visitors_b_daily, left_on='date', right_on='date', how='left')\
    .merge(visitors_a_cum, left_on='date', right_on='date', how='left')\
    .merge(visitors_b_cum, left_on='date', right_on='date', how='left')
data_total.head()
```

Названия столбцов в таблице:

- date — дата;
- orders_per_date_a — количество заказов в выбранную дату в группе A;
- revenue_per_date_a — суммарная выручка в выбранную дату в группе A;
- orders_per_date_b — количество заказов в выбранную дату в группе B;
- revenue_per_date_b — суммарная выручка в выбранную дату в группе B;
- orders_cum_a — суммарное число заказов до выбранной даты включительно в группе A;
- revenue_cum_a — суммарная выручка до выбранной даты включительно в группе A;
- orders_cum_b — суммарное количество заказов до выбранной даты включительно в группе B;
- revenue_cum_b — суммарная выручка до выбранной даты включительно в группе B;
- visitors_per_date_a — количество пользователей в выбранную дату в группе A;
- visitors_per_date_b — количество пользователей в выбранную дату в группе B;
- visitors_a_cum — количество пользователей до выбранной даты включительно в группе A;
- visitors_b_cum — количество пользователей до выбранной даты включительно в группе B.

Создадим переменные orders_users_a и orders_users_b. В них для пользователей, которые заказывали хотя бы 1 раз, укажем число совершённых заказов.

```python
orders_users_a = orders[orders['group']=='A'].groupby('visitor_id', as_index=False).agg({'transaction_id' : pd.Series.nunique})
orders_users_a.columns = ['visitor_id', 'orders']

orders_users_b = orders[orders['group']=='B'].groupby('visitor_id', as_index=False).agg({'transaction_id' : pd.Series.nunique})
orders_users_b.columns = ['visitor_id', 'orders']
```

Объявим переменные sample_a и sample_b, в которых пользователям из разных групп будет соответствовать количество заказов. Тем, кто ничего не заказал, будут соответствовать нули. Это нужно, чтобы подготовить выборки к проверке критерием Манна-Уитни.

Структура переменнной sample_a:
- Список с количеством заказов для каждого из пользователей: orders_users_a['orders'].
- Нули для пользователей, которые ничего не заказывали. Их количество равно разнице между суммой посетителей и количеством записей о заказах: data['visitors_per_date_a'].sum() - len(orders_users_a['orders']). Создадим объект pd.Series нужной длины:

```python
sample_a = pd.concat(
    [orders_users_a['orders'],pd.Series(0, index=np.arange(data_total['visitors_per_date_a'].sum() - len(orders_users_a['orders'])), \
                                        name='orders')], axis=0)

sample_b = pd.concat(
    [orders_users_b['orders'],pd.Series(0, index=np.arange(data_total['visitors_per_date_b'].sum() - len(orders_users_b['orders'])), \
                                        name='orders')], axis=0)
```

**Нулевая гипотеза:** среднее количество заказов на посетителя по «сырым» данным у групп A и B одинаковое

**Альтернативная гипотеза:** среднее количество заказов на посетителя по «сырым» данным у групп A и B различается.

```python
#Критический уровень статистической значимости
alpha = .05 

results = stats.mannwhitneyu(sample_a, sample_b)
print('p-значение: ', "{0:.3f}".format(results.pvalue))

if (results.pvalue < alpha):
    print('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print('Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя')

print("{0:.3f}".format(sample_b.mean()/sample_a.mean()-1))
```

**Итог:** среднее количество заказов на посетителя по «сырым» данным у групп A и B различается - у группы B значение выше на 13.8%.

### Статистическая значимость различий в среднем чеке заказа между группами по «сырым» данным

**Нулевая гипотеза:** средний чек заказа групп A и B по «сырым» данным одинаковый.

**Альтернативная гипотеза:** средний чек заказа групп A и B по «сырым» данным различается.

Рассчитаем статистическую значимость различий в среднем чеке между сегментами, передадав критерию mannwhitneyu() данные о выручке с заказов. Найдём относительные различия в среднем чеке между группами.

```python
results = stats.mannwhitneyu(orders[orders['group']=='A']['revenue'], orders[orders['group']=='B']['revenue'])

print('p-значение: ', "{0:.3f}".format(results.pvalue))

if (results.pvalue < alpha):
    print('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print('Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя')

print("{0:.3f}".format(orders[orders['group']=='B']['revenue'].mean()/orders[orders['group']=='A']['revenue'].mean()-1))
```

**Итог:** средний чек заказа по «сырым» данным у групп A и B не различается.

### Статистическая значимость различий в среднем количестве заказов на посетителя по «очищенным» данным

Сделаем срезы пользователей с числом заказов больше 4 — users_with_many_orders и пользователей, совершивших заказы дороже 60 000 — users_with_expensive_orders. Объединим их в таблице abnormal_users. Выведем их количество.

```python
many_orders = np.percentile(orders_by_users['transactions'], 99)
expensive_orders = np.percentile(orders['revenue'], 99)
users_with_many_orders = pd.concat(
    [orders_users_a[orders_users_a['orders'] > many_orders]['visitor_id'], \
     orders_users_b[orders_users_b['orders'] > many_orders]['visitor_id']], axis = 0)
users_with_expensive_orders = orders[orders['revenue'] > expensive_orders]['visitor_id']
abnormal_users = pd.concat([users_with_many_orders, users_with_expensive_orders], axis = 0).drop_duplicates().sort_values()
abnormal_users.shape
```

В сумме имеем 15 аномальных пользователей.

Проанализируем, как их действия повлияли на результаты теста. Посчитаем статистическую значимость различий в среднем количестве заказов на посетителя по «очищенным» данным. Сначала подготовим выборки количества заказов по пользователям по группам теста.

```python
sample_a_clean = pd.concat(
    [orders_users_a[np.logical_not(orders_users_a['visitor_id'].isin(abnormal_users))]['orders'], \
     pd.Series(0, index=np.arange(data_total['visitors_per_date_a'].sum() - len(orders_users_a['orders'])),name='orders')],axis=0)

sample_b_clean = pd.concat(
    [orders_users_b[np.logical_not(orders_users_b['visitor_id'].isin(abnormal_users))]['orders'], \
     pd.Series(0, index=np.arange(data_total['visitors_per_date_b'].sum() - len(orders_users_b['orders'])),name='orders')],axis=0)
```

**Нулевая гипотеза:** среднее количество заказов на посетителя по «очищенным» данным у групп A и B одинаковое

**Альтернативная гипотеза:** среднее количество заказов на посетителя по «очищенным» данным у групп A и B различается.

```python
results = stats.mannwhitneyu(sample_a_clean, sample_b_clean)

print('p-значение: ', '{0:.3f}'.format(results.pvalue))

if (results.pvalue < alpha):
    print('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print('Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя')

print('{0:.3f}'.format(sample_b_clean.mean()/sample_a_clean.mean()-1))
```

**Итог:** значение группы B все равно лучше группы A, значит наличие аномалий не повлияло на результат.

### Статистическая значимость различий в среднем чеке заказа между группами по «очищенным» данным

**Нулевая гипотеза:** средний чек заказа групп A и B по «очищенным» данным одинаковый.

**Альтернативная гипотеза:** средний чек заказа групп A и B по «очищенным» данным различается.

```python
results = stats.mannwhitneyu(
    orders[np.logical_and(
        orders['group']=='A',
        np.logical_not(orders['visitor_id'].isin(abnormal_users)))]['revenue'],
    orders[np.logical_and(
        orders['group']=='B',
        np.logical_not(orders['visitor_id'].isin(abnormal_users)))]['revenue'])

print('p-значение: ', '{0:.3f}'.format(results.pvalue))

if (results.pvalue < alpha):
    print('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print('Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя')

print('{0:.3f}'.format(
    orders[np.logical_and(orders['group']=='B',np.logical_not(orders['visitor_id'].isin(abnormal_users)))]['revenue'].mean()/
    orders[np.logical_and(
        orders['group']=='A',
        np.logical_not(orders['visitor_id'].isin(abnormal_users)))]['revenue'].mean() - 1))
```

**Итог:** средний чек заказа по «сырым» данным у групп A и B не различается, результат теста тот же, значит наличие аномалий не повлияло на него.

### Решение по результатам теста

Сравним кумулятивный среднего чека по группам по «очищенным» данным. Сохраним кумулятивные данные о заказах в таблицах orders_a_clean и orders_b_clean.

```python
orders_a_clean = orders[np.logical_and(
        orders['group']=='A',
        np.logical_not(orders['visitor_id'].isin(abnormal_users)))][['date', 'transaction_id', 'revenue']].groupby(
    ['date']).agg({'transaction_id':'nunique', 'revenue':'sum'}).cumsum().reset_index()

orders_b_clean = orders[np.logical_and(
        orders['group']=='B',
        np.logical_not(orders['visitor_id'].isin(abnormal_users)))][['date', 'transaction_id', 'revenue']].groupby(
    ['date']).agg({'transaction_id':'nunique', 'revenue':'sum'}).cumsum().reset_index()
```

```python
plt.figure(figsize=(10,5))
plt.grid(True)
plt.title('График среднего чека по очищенным данным по дням и группам A/B-тестирования')

plt.plot(orders_a_clean['date'], orders_a_clean['revenue'] / orders_a_clean['transaction_id'], label='A')
plt.plot(orders_b_clean['date'], orders_b_clean['revenue'] / orders_b_clean['transaction_id'], label='B')

ax = plt.gca()
ax.grid(True)
ax.set_xlabel('Дни заказов')
ax.set_ylabel('Средний чек заказов')
plt.legend()
```

Построим график относительного изменения кумулятивного среднего чека по «очищенным» данным.

```python
revenue_cum_clean = orders_a_clean.merge(
    orders_b_clean, left_on='date', right_on='date', how='left', suffixes=['A', 'B'])
```

```python
plt.figure(figsize=(10, 7))
plt.grid(True)
plt.title('График относительного изменения кумулятивного среднего чека группы B к группе A по очищенным данным')
plt.plot(revenue_cum_clean['date'], \
         (revenue_cum_clean['revenueB']/revenue_cum_clean['transaction_idB'])/ \
         (revenue_cum_clean['revenueA']/revenue_cum_clean['transaction_idA'])-1)

ax = plt.gca()
ax.set_xlabel('Дни заказов')
ax.set_ylabel('Относительное изменение кумулятивного среднего чека')

# Добавляем нулевую ось X
plt.axhline(y=0, color='black', linestyle='dotted', linewidth = 3)
```

Есть статистически значимое различие по количеству заказов на посетителя между группами как по «сырым» данным, так и по «очищенным». График различия конверсии между группами показывает, что результаты группы B лучше группы A.

Нет статистически значимого различия по среднему чеку между группами ни по «сырым» данным, ни по «очищенным».
