### Работа с пустыми значениями, замена типов данных

```python
#Определим в каких столбцах есть пропуски и их долю
print(data.isna().sum().sort_values(ascending=False))
print('Процент пустых значений по столбцам:')
print(pd.DataFrame(round(data.isna().mean()*100,)))
```

#### Столбцы name, genre - название и жанр

```python
#Удалим пустоты в столбцах  name, genre, их всего по 2 шт.
data = data.drop(np.where(data['name'].isna())[0])
data = data.drop(np.where(data['genre'].isna())[0])
```

#### Столбец year_of_release - год выпуска

```python
#Удалим пустоты в столбце year_of_release, они составляют незначительную долю - 2%
data = data.drop((data[data['year_of_release'].isna()].index), axis=0) 
```

```python
#Заменим типа данных на int
data['year_of_release'] = data['year_of_release'].astype(int)
```

#### Столбец critic_score - оценка критиков

```python
#Изучим какие данные есть в столбцах critic_score

#Функция для вывода информации о данных в столбце
def total_info_data(data): 
    print('Уникальные значения столбца', data.name, ':', data.unique())
    print('Количество пропусков столбца', data.name, ':', data.isna().sum())
    print('Основные статистические данные: \n', data.describe())
    
def hist_data(data):     
    fig, ax = plt.subplots(figsize=(6, 3))
    n_bins = len(data.unique())
    ax.hist(data, bins=n_bins)
    ax.set_title('Распределение данных в столбце')
    ax.set_xlabel(data.name)
    ax.set_ylabel('games')

total_info_data(data['critic_score'])
hist_data(data['critic_score'])
```

Данные в столбце critic_score означают оценку критиков, где максимальное значение равно 100. 
Например, в одном из самых популярных сайтов-агрегаторов, собирающих  отзывы [Metacritic](https://www.metacritic.com/faq) оценка игры расчитывается как средневзвешенное значение отзывов ведущих критиков и публикаций. Metacritic отбирает игровые издания, сопоставляет их рецензии, трансформирует их оценки в показатели по шкале от 0 до 100, после чего на их основе вычисляет единую оценку для каждой игры.

Есть несколько вариантов заполнения пустот в этом столбце, но все они имеют свои минусы:
1) Можно было бы предположить, что critic_score зависит от продаж, и проставить вместо пустот медианное значение согласно уровню продаж. _НО даже при средней корреляции совсем не объективно делать вывод, что низкая оценка критиков может привести к падению продаж и наоборот. Такой способ замены пустот создаст искусственную зависимость, что исказит итоги исследования._

2) Можно было бы вытащить оценки из Metacritic или иного сайта и заменить пустоты на рейтинг оттуда. _НО мы не знаем достоверно источник и дату выгрузки данных нашего датасета, поэтому замена пустот на рейтинг сегодняшнего дня также может исказить наше исследование._

3) Возможно отсутствие данных означало отсутствие рейтинга, например, если игры совсем новые или очень старые. Тогда пустоты можно было бы заменить на ноль. _НО отсутствие critic_score наблюдается у игр, вышедших на протяжении всего временного ряда. Поэтому причина появления пропусков все еще остается не понятным._  

**В итоге принято решение в данном столбце оставить пустоты как есть.**

#### Столбец user_score - оценка пользователей

```python
#Изучим какие данные есть в столбцах user_score
total_info_data(data['user_score'])
```

В этом столбце сразу 3 проблемы - неподходящий тип, пустоты и значние tbd.

- Тип данных нужно заменить на float.

- В целях дальнейшей замены типа данных на float нам нужно сделать пустоты числовыми. Заменим временно пустоты на -1

- В столбце есть текстовое значение tbd, что означает To Be Determined, то есть "Будет определено". В контексте оценок пользователей, tbd означает, что оценок еще слишком мало или они отсутсвуют, поэтому user_score пока не может быть определен. Соответственно, логично заменить это значение на пустоты (но сначала временно на -1).

Как и в столбце critic_score, мы не можем заменить (по перечисленным в предыдущем подразделе причинам) пустоты на нули, внешние данные или медианы с условием. 

**Принято решение в данном столбце оставить пустоты как есть, но нужно сделать заменить их сначала на -1, потом снова сделать  их nan.**

```python
#В целях дальнейшей замены типа данных на float нам нужно заменить текст на числовое значение. Заменяем tbd на -1
data['user_score'] = data['user_score'].str.replace('tbd','-1')
```

```python
#В целях дальнейшей замены типа данных на float нам нужно сделать пустоты числовыми. Заменим пустоты на -1
data['user_score'] = data['user_score'].fillna(-1)

#Заменим тип данных на float
data['user_score'] = pd.to_numeric(data['user_score'] , downcast='float')
```

```python
#Возвращаем пустоты на место, заменяем -1 на nan
data.loc[data['user_score'] == -1, 'user_score'] = np.nan
```

```python
#Заново смотрим какие данные есть в столбцах user_score
total_info_data(data['user_score'])
hist_data(data['user_score'])
```

Итак, в столбце user_score тип данных заменили на float.
Ячейки со значением tbd сделали пустыми (оценок еще слишком мало или они отсутсвуют).
Первоначальные пустоты оставили на месте.

#### Столбец rating - рейтинг от организации ESRB 

```python
#Изучим какие данные есть в столбце rating

total_info_data(data['rating'])
print(data['rating'].value_counts())
```

ESRB (англ. Entertainment Software Rating Board) - ассоциация, которая определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию. 

>EC	- Для детей младшего возраста. Подходит для детей от 3 лет, неприемлемое содержание отсутствует. Примеры: Дора Исследовательница, Сказки Дракона.

>E	- Подходит для всех возрастных категорий. Игра не должна содержать звуков или изображений, которые могут напугать маленьких детей. Не должна использоваться ненормативная лексика. Примеры: Просто таннцуй, FIFA.

> E 10+	- Подходит для лиц старше 10 лет. Игра может содержать умеренные формы насилия, а некоторые сцены могут пугать детей. Примеры: Minecraft Dungeons, Растения против зомби.

>T	- Подходит для лиц старше 13 лет. Игра может содержать более реалистичные и натуралистические сцены насилия. Примеры: Fortnite, Sims 4.

>M	- Подходит для лиц старше 17 лет. Этот рейтинг используется, когда насилие становится реалистичным и похожим на реальную жизнь. Также может присутствовать ненормативная лексика, употребление табака, алкоголя или запрещенных наркотиков. Примеры: Ark: Survival Evolved, Destiny 2.

>AO	- Подходит для взрослых старше 18 лет. Классификация «для взрослых» используется, когда в игре присутствует крайний уровень насилия и немотивированного убийства. Также может присутствовать романтизация наркотиков, азартных игр и сексуальной активности. Примеры: Grand Theft Auto V, Fallout 4.

>RP	- Категория ещё не присвоена. Играм с категорией RP еще не присвоен окончательный рейтинг ESRB.

Можно было бы определить возрастной рейтинг по жанру игры, но игры одного жанра могут иметь разный контент, влияющий на возратсные ограничения.

В данном случае пустоты можно заменить двумя способами:
- Замена пустот на RP - категория рейтинга еще не присвоена. НО в играх с пустотами в столбце с рейтингом годы выпуска находятся в промежутке 1981-2016 гг.
- Замена пустот на unknown

#### Как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков

Проанализируем, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков.

```python
# Функция для вывода графиков   
def corr_func(platform):
    temp = actual_data.query('platform ==@platform\
                                and user_score > 0\
                                and critic_score > 0').reset_index(drop=True)

    plt.figure(figsize=(16,4))
    plt.subplot(1, 3, 1)
    sns.heatmap(temp.loc[:,['sale_total', 'user_score', 'critic_score']].corr(),annot=True);      
    plt.subplot(1, 3, 2)
    sns.scatterplot(x='user_score',y='sale_total',hue = 'critic_score',size = 'critic_score',sizes = (20,200), data=temp);
    plt.subplot(1, 3, 3)
    sns.scatterplot(x='critic_score',y='sale_total',hue = 'user_score',size = 'user_score',sizes = (20,200), data=temp);

# Проверим функцию на платформе X360   
corr_func('X360')
```

Для платформы X360 влияние отзывов на продажи следующее:
- Отзывы пользователей не влияют на общие продажи (минусовая корреляция -0.0057)
- Отзывы критиков влияют средне (положительная корреляция 0.35)

```python
# Применим функцию корреляции на платформе XOne   
corr_func('XOne')
```

Для платформы XOne влияние отзывов на продажи следующее:

- Отзывы критиков влияют средне (положительная корреляция 0.41), но сильнее, чем у X360

```python
# Применим функцию корреляции на платформе PS4   
corr_func('PS4')
```

Для платформы PS4 влияние отзывов на продажи следующее:

### Проверка гипотез

Проверим две гипотезы:

1) Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

2) Средние пользовательские рейтинги жанров Action и Sports разные

- Зададим пороговое значение alpha равное 0.01

- Возьмем актуальный период времени 2013-2016 гг.

- Сформулируем нулевую и альтернативную гипотезы

- Гипотезы проверим по методу ttest_ind – T-тест для средних двух независимых выборок баллов. Это двусторонний тест для нулевой гипотезы о том, что две независимые выборки имеют идентичные средние (ожидаемые) значения.

#### Гипотеза № 1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

H0 - Нулевая гипотеза: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

H1 - Альтернативная гипотеза: Средние пользовательские рейтинги платформ Xbox One и PC отличаются


```python
# Сохраним в переменных xbox_us и pc_us соответствующие данные (за 2013-2016 гг.)
xone_us = fresh_data.loc[(data['platform']=='XOne')&(fresh_data['user_score']!=0)]['user_score']
pc_us = fresh_data.loc[(data['platform']=='PC')&(fresh_data['user_score']!=0)]['user_score']
```

```python
#Выполним проверку гипотезы по методу ttest_ind
alpha = .01
results = st.ttest_ind(
    xone_us.dropna(), 
    pc_us.dropna(),
    equal_var = False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print('Отвергаем нулевую гипотезу, рейтинги платформ Xbox One и PC разные')
else:
    print('Не получилось отвергнуть нулевую гипотезу, рейтинги платформ Xbox One и PC одинаковые')
```

При проведении статистического теста выяснилось, что рейтинги платформ Xbox One и PC одинаковые, нулевая гипотеза не отвергнута. 

#### Гипотеза № 2. Средние пользовательские рейтинги жанров Action и Sports разные

H0 - Нулевая гипотеза: Средние пользовательские рейтинги жанров Action и Sports одинаковые

H1 - Альтернативная гипотеза: Средние пользовательские рейтинги жанров Action и Sports разные

```python
# Сохраним в переменных action_us и sports_us соответствующие данные  (за 2013-2016 гг.)
action_us = fresh_data.loc[(data['genre']=='Action')&(fresh_data['user_score']!=0)]['user_score']
sports_us = fresh_data.loc[(data['genre']=='Sports')&(fresh_data['user_score']!=0)]['user_score']
```

```python
#Выполним проверку гипотезы по методу ttest_ind

alpha = .01
results = st.ttest_ind(
    action_us.dropna(), 
    sports_us.dropna(),
    equal_var = False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print('Отвергаем нулевую гипотезу, рейтинги рейтинги жанров Action и Sports разные')
else:
    print('Не получилось отвергнуть нулевую гипотезу, рейтинги  жанров Action и Sports одинаковые')
```

При проведении статистического теста по всему периоду времени, выяснилось, что рейтинги жанров Action и Sports разные и нулевая гипотеза отвергнута.

- Гипотеза №2: Средние пользовательские рейтинги жанров Action и Sports разные. Выяснилось, что рейтинги жанров Action и Sports разные и нулевая гипотеза отвергнута.
