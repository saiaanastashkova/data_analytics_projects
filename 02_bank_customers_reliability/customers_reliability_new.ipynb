# Исследование надежности заемщиков

Заказчик — кредитный отдел банка. Нужно разобраться, влияет ли семейное положение и количество детей клиента на факт погашения кредита в срок. Входные данные от банка — статистика о платёжеспособности клиентов.

 4. Выводы и рекомендации для учета при построении модели кредитного скоринга.

### Устранение пропусков

В датасете есть пропуски всего в двух столбцах - days_employed (общий трудовой стаж) и total_income (ежемесячный доход). Число пропущенных значений в первом и втором столбце равные, по 2174 в каждом. Также пропуски в этих слолбцах появляются одновременно. При этом портреты клиентов по остальным характеристикам (пол, возраст, образование, количество детей и др.) самые разнообразные.

Поэтому действительное отсутствие стажа / дохода, невнимательность при заполнении полей не могут рассматриваться как возможные причины появления пропусков.

**Итак, пропуски могли появиться по следующим причинам**:
- Ошибка программы при вводе данных
- Ошибка сервера при выгрузке данных

В случае необходимости заполнения пропусков для количественных переменных есть несколько решений, и все они имеют свои особенности:
- **Заполнение нулем** - может сильно исказить анализ, если проставить их безосновательно. Может применяться, только если действительно есть основания считать, что пропуски означают нулевой показатель. Например, если индивидуальный предприниматель не заполнил поле с количеством сотрудников, потому что у него их 0.
- **Заполнение средним** - также может быть необъективным, особенно если есть аномальные значения или если амплитуда данных слишком большая (_"У Пети 10 яблок, а у Васи —  0. В среднем у каждого по 5 яблок"_).
- **Заполнение медианой** - наиболее оптимальное решение, т.к. она более "справдлива", устойчива к аномальным значениям и "выбросам" данных. 

При заполнении медианным значением при возможности нужно отфильтровать категорию по наиболее влиятельным показателям. Так и было сделано в текущем проекте - пропуски в уровне дохода заполнены медианой по типу занятости.

На сумму дохода сильнее всего влияет тип занятости, поэтому заполнить пропуски в этом столбце нужно медианным значением по каждому типу из столбца income_type. Например, у человека с типом занятости сотрудник пропуск в столбце total_income должен быть заполнен медианным доходом среди всех записей с тем же типом.

```python
# Заполнение пустот в столбцах income_type и total_income медианным доходом по типу занятости
for t in data['income_type'].unique():
    data.loc[(data['income_type'] == t) & (data['total_income'].isna()), 'total_income'] = \
    data.loc[(data['income_type'] == t), 'total_income'].median()
```

### Изучение аномалий

В данных могут встречаться артефакты (аномалии) — значения, которые не отражают действительность и появились по какой-то ошибке. таким артефактом будет отрицательное количество дней трудового стажа в столбце days_employed. 

```python
# Заменим все отрицательные значения положительными с помощью метода abs()
data['days_employed'] = data['days_employed'].abs()
```

Для каждого типа занятости выведем медианное значение трудового стажа days_employed в днях.

```python
data.groupby('income_type')['days_employed'].agg('median')
```

У двух типов (безработные и пенсионеры) получатся аномально большие значения. Исправить такие значения сложно, поэтому оставим их как есть. Тем более этот столбец не понадобится вам для исследования.

```python
# Выведем перечень уникальных значений столбца children.
data['children'].unique()
```

В столбце children есть два аномальных значения. Удалим строки, в которых встречаются такие аномальные значения.

```python
# Удалим строки с аномальными значениями
data = data[(data['children'] != -1) & (data['children'] != 20)]
```

Заполним пропуски в столбце c трудовым стажем медианными значениями по каждого типа занятости.

```python
# Заполним пропуски в столбце days_employed медианными значениями по каждого типа занятости income_type 
for t in data['income_type'].unique():
    data.loc[(data['income_type'] == t) & (data['days_employed'].isna()), 'days_employed'] = \
    data.loc[(data['income_type'] == t), 'days_employed'].median()
```

### Изменение типов данных

Заменим тип данных в столбце total_income на целочисленный.

```python
# Изменим тип данных в столбце total_income на int
data['total_income'] = data['total_income'].astype(int)
```

### Обработка дубликатов

Обработаем неявные дубликаты в столбце education. В этом столбце есть одни и те же значения, но записанные по-разному: с использованием заглавных и строчных букв. Приведем их к нижнему регистру. 

```python
# Приведем к нижнему регистру значения столбца education
data['education'] = data['education'].str.lower()
```

Посмотрим остались ли явные дубликаты после всех проведенных манипуляций.

```python
# Количество строк-дубликатов в данных
data.duplicated().sum()
```

```python
# Удаление строк-дубликатов
data = data.drop_duplicates()
```

### Категоризация данных

На основании диапазонов, указанных ниже, создадим в датафрейме `data` столбец `total_income_category` с категориями:

- 0–30000 — `'E'`;
- 30001–50000 — `'D'`;
- 50001–200000 — `'C'`;
- 200001–1000000 — `'B'`;
- 1000001 и выше — `'A'`.


Например, кредитополучателю с доходом 25000 нужно назначить категорию `'E'`, а клиенту, получающему 235000, — `'B'`.

```python
# Функция категоризации доходов
def categorize_income(income):
    try:
        if 0 <= income <= 30000:
            return 'E'
        elif 30001 <= income <= 50000:
            return 'D'
        elif 50001 <= income <= 200000:
            return 'C'
        elif 200001 <= income <= 1000000:
            return 'B'
        elif income >= 1000001:
            return 'A'
    except:
        pass
```

```python
# Используем функцию категоризации доходов
data['total_income_category'] = data['total_income'].apply(categorize_income)
```

Рассмотрим что указано в качестве уникальных целей взятия кредита.

```python
# Перечень уникальных целей взятия кредита из столбца purpose
data['purpose'].unique()
```

Создадим функцию, которая на основании данных из столбца `purpose` сформирует новый столбец `purpose_category`, в который войдут следующие категории:

- `'операции с автомобилем'`,
- `'операции с недвижимостью'`,
- `'проведение свадьбы'`,
- `'получение образования'`.

Например, если в столбце `purpose` находится подстрока `'на покупку автомобиля'`, то в столбце `purpose_category` должна появиться строка `'операции с автомобилем'`.

```python
# Функция категоризации целей взятия кредита
def categorize_purpose(row):
    try:
        if 'автом' in row:
            return 'операции с автомобилем'
        elif 'жил' in row or 'недвиж' in row:
            return 'операции с недвижимостью'
        elif 'свад' in row:
            return 'проведение свадьбы'
        elif 'образов' in row:
            return 'получение образования'
    except:
        return 'нет категории'
```

```python
# Используем функцию категоризации целей взятия кредита
data['purpose_category'] = data['purpose'].apply(categorize_purpose)
```

## Исследовательский анализ данных

### Есть ли зависимость между количеством детей и возвратом кредита в срок

```python
# Сначала посмотрим как часто берут кредиты люди с разным количеством детей.
print(data['children'].value_counts())
```

```python
# Проверим количество клиентов, имеющих задолженности по кредитам, методом группировки по количеству детей.
# Отсортируем данные по убыванию.
print(data.groupby('children')['debt'].sum().sort_values(ascending=False))

# Выведем долю должников о общего числа клиентов с определенным количеством детей через среднюю.
# Отсортируем данные по убыванию.
print(data.groupby('children')['debt'].mean().sort_values(ascending=False))
```

**Вывод:** Сначала может показаться, что должников больше всего среди основного клиентского сегмента банка - у людей без детей. Но если сравнить долю должников, то эта категория наоборот наиболее надежная _(клиентов с 5 детьми не считаем, их всего около 0,04% от общего числа)_. 
Зависимость между количеством детей и возвратом кредита в срок есть: **люди с детьми чаще имеют задолженности**. При этом точное количество детей может влиять по-разному - доля должников с 3 детьми меньше, чем доля должников с 2 детьми.

### Есть ли зависимость между семейным положением и возвратом кредита в срок

```python
# Проверим зависимость тем же способом, как в предыдущем пункте.

# Сначала посмотрим как часто берут кредиты люди с разным семейным статусом.
print(data['family_status'].value_counts())

# Проверим количество клиентов, имеющих задолженности.
# Отсортируем данные по убыванию.
print(data.groupby('family_status')['debt'].sum().sort_values(ascending=False))

# Выведем долю должников через среднюю.
# Отсортируем данные по убыванию.
print(data.groupby('family_status')['debt'].mean().sort_values(ascending=False))
```

**Вывод:** Зависимость между семейным положением и возвратом кредита в срок есть: **люди, не состоящие (и не состоявшие ранее) в официальном браке, чаще имеют задолженности**. Наиболее надежной категорией лиц являются вдовцы / вдовы, наименее надежной - не женатые / не замужние, а также состоящие в гражданском браке. 

### Есть ли зависимость между уровнем дохода и возвратом кредита в срок

```python
# Проверим зависимость тем же способом, как в предыдущих пунктах.

# Сначала посмотрим как часто берут кредиты люди с разным уровнем доходов. Обратимся к столбцу с категориями дохода A-E.
# - 0–30000 — `'E'`;
# - 30001–50000 — `'D'`;
# - 50001–200000 — `'C'`;
# - 200001–1000000 — `'B'`;
# - 1000001 и выше — `'A'`.
print(data['total_income_category'].value_counts())

# Проверим количество клиентов, имеющих задолженности.
# Отсортируем данные по убыванию.
print(data.groupby('total_income_category')['debt'].sum().sort_values(ascending=False))

# Выведем долю должников через среднюю.
# Отсортируем данные по убыванию.
print(data.groupby('total_income_category')['debt'].mean().sort_values(ascending=False))
```

**Вывод:**  Между уровнем дохода и возвратом кредита в срок **не наблюдается прямой зависимости**. Если убрать из списка людей с доходом A (1000001 и выше) и E (0–30000) люди, составляющих около 0,1% от общего числа, то наименее надежными являются клиенты с доходом C (50001–200000 руб.), далее - клиенты с доходом B (200001–1000000 руб.), а наиболее - клиенты с доходом D (30001–50000 руб.)


### Как разные цели кредита влияют на его возврат в срок

```python
# Проверим зависимость тем же способом, как в предыдущих пунктах.

# Сначала посмотрим как часто берут кредиты люди с разными целями на кредит. Обратимся к столбцу с категориями.
print(data['purpose_category'].value_counts())

# Проверим количество клиентов, имеющих задолженности.
# Отсортируем данные по убыванию.
print(data.groupby('purpose_category')['debt'].sum().sort_values(ascending=False))

# Выведем долю должников через среднюю.
# Отсортируем данные по убыванию.
print(data.groupby('purpose_category')['debt'].mean().sort_values(ascending=False))
```

**Вывод:** **Наиболее ответственно** люди относятся к выплате самого популярного вида кредитов - **на операции с недвижимостью**. Задолженности чаще всего допускаются в случаях, когда кредит был взят на операции с автомобилем.

Зависимость между количеством детей и возвратом кредита в срок есть: **люди с детьми чаще имеют задолженности, чем люди без детей**. То есть на возврат крелита в срок влияет именно факт наличия или отсутствия детей, при этом по имеющимся данным пока нельзя сделать вывод по влиянию точного количества детей. Для детального анализа нужна более репрезентативная выборка, чтобы исследуемые группы были близких размеров.

-------------    | -------------
4                |  0.097561
2                |  0.094542
1                |  0.092346
3                |  0.081818
нет детей        |  0.075438

**Гипотеза 2.** Подтверждена.
Зависимость между семейным положением и возвратом кредита в срок есть: **люди, не состоящие (и не состоявшие ранее) в официальном браке, чаще имеют задолженности**. Наименее надежной - не женатые / не замужние, а также состоящие в гражданском браке. Наиболее надежной категорией лиц являются вдовцы / вдовы, но их доля среди клиентов банка довольная маленькая.

-------------          | -------------
не женат / не замужем  |  0.097639
гражданский брак       |  0.093130
женат / замужем        |  0.075606
в разводе              |  0.070648
вдовец / вдова         |  0.066246

**Гипотеза 3.** Частично подтверждена. 
Между уровнем дохода и возвратом кредита в срок **не наблюдается прямой зависимости**. Если убрать из списка людей с доходом A (1000001 и выше) и E (0–30000) люди, составляющих около 0,1% от общего числа, то наименее надежными являются клиенты с доходом C (50001–200000 руб.), далее - клиенты с доходом B (200001–1000000 руб.), а наиболее - клиенты с доходом D (30001–50000 руб.). Для подробного анализа нужна более широкая выборка данных по клиентам с доходами 0–30000 руб., от 1000001 руб. и выше.
Также следует обратить внимание, что 10% данных были заполнены искусственно медианным значением.

Уровень дохода         | Доля должников
-------------          | -------------
C (50001–200000 руб.)  | 0.084982
B (200001–1000000 руб.)|0.070602
D (30001–50000 руб.)   |0.060172

**Гипотеза 4.** Подтверждена. 
**Наиболее ответственно** люди относятся к выплате самого популярного вида кредитов - **на операции с недвижимостью**. Задолженности чаще всего допускаются в случаях, когда кредит был взят на операции с автомобилем.
